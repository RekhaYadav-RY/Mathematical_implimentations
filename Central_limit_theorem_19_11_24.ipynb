```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stat
```

```python
# Assume that
# X1 - Uniform continuous random variable.
# X2 - Bernoulli random variable.
# X3 - Binomial discrete random variable.
# X4 - Geometric random variable.
# X5 - Poisson discrete random variable.

from scipy.stats import uniform, bernoulli, binom, geom, poisson
```

```python
# Let us take 250 samples from the above-mentioned distributions of size 1000.
# And store these value in an array of size (250, 1000) where each row represents a sample of size 1000.
No_sample = 250
Sample_size = 1000
                                                                    # Here mean = population mean, variance = population variance
X1 = uniform.rvs(size=(No_sample, Sample_size))  #a = 0, b = 1      # mean = a+b/2 = 1/2, variance = (b-a)^2/12 = 1/12
X2 = bernoulli.rvs(p = 0.3, size = (No_sample, Sample_size))        # mean = p = 0.3, variance = p(1-p) = 0.21
X3 = binom.rvs(n = 6, p = 0.4, size=(No_sample, Sample_size))       # mean = np = 2.4, variance = np(1-p) = 1.44
X4 = geom.rvs(p = 0.5, size=(No_sample, Sample_size))               # mean = 1/p = 2, variance = (1-p)/p^2 = 2
X5 = poisson.rvs(mu = 0.6, size=(No_sample, Sample_size))           # mean = mu = 0.6, variance = mu = 0.6
```

```python
X1_mean = np.mean(X1, axis=1)     # Sample mean of X1
X2_mean = np.mean(X2, axis=1)
X3_mean = np.mean(X3, axis=1)
X4_mean = np.mean(X4, axis=1)
X5_mean = np.mean(X5, axis=1)
```

### Central Limit Theorem
### sample mean of S = population mean
### sample standard deviation of S = population standard deviation /SQRT(no. of sample)
### As per the Central Limit Theorem, the sample mean S will approximate to the normal distribution which is given as S ⁓ N(μ, σ/√n).

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
sns.histplot(X1_mean, bins=30, kde=True, stat="density")
plt.title('Uniform Distribution')
plt.subplot(2, 3, 2)
sns.histplot(X2_mean, bins=30, kde=True, stat="density")
plt.title('Bernoulli Distribution')
plt.subplot(2, 3, 3)
sns.histplot(X3_mean, bins=30, kde=True, stat="density")
plt.title('Binomial Distribution')
plt.subplot(2, 3, 4)
sns.histplot(X4_mean, bins=30, kde=True, stat="density")
plt.title('Geometric Distribution')
plt.subplot(2, 3, 5)
sns.histplot(X5_mean, bins=30, kde=True, stat="density")
plt.title('Poisson Distribution')
plt.show()
```
